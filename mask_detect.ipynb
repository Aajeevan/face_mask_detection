```python
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import numpy
```

```python
face_cas = cv2.CascadeClassifier('haarcascade_frontalface_alt2 (1).xml')
face_cas
```

```python
model = load_model('mask_recog.h5')
```

```python
def face_mask_detection(frame):
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_cas.detectMultiScale(gray,minNeighbors=4)
    print(faces)
    for x,y,w,h in faces:
        face_frame = frame[y:y+h,x:x+w]
        #print(face_frame.shape)
    face_frame=cv2.cvtColor(face_frame, cv2.COLOR_BGR2RGB)
    face_frame=cv2.resize(face_frame,(224,224))
    face_frame=face_frame.reshape(1,224, 224,3)
    #print(face_frame)
    face_frame=preprocess_input (face_frame)
    #print(face_frame)
    pred=model.predict(face_frame)[0]
    print(pred)
    with_mask,without_mask=pred
    label="Mask" if with_mask>without_mask else "No mask"
    print(label)
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),3)
    cv2.putText(frame,label,(x,y),cv2.FONT_HERSHEY_COMPLEX,2,(255,0,0),2)
    return frame
```

```python
img = cv2.imread('nomask.jpg')
result=face_mas_detection(img)
cv2.imshow("RESULT", result)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Just Face detection
video = cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(gray_img)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,2,34),3)
    cv2.imshow("WEBCAM",img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
